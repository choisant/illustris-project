# Faber-Jackson relation
---

The Faber-Jackson relation gives a relation between the luminosity (or total mass) of a galaxy and the velocity dispersion of its stars. This relation is a power law:
$$L \propto \sigma^{\gamma} $$
The exponent $\gamma$ is approximately 4. 

To compare our data with this power law, the (1D) velocity dispersion of each subhalo (satellite galaxy) is plotted as a function of the total mass of that subhalo.



```python
import illustris_python as il
import matplotlib.pyplot as plt
import numpy as np
```

```python
basePath = "../data/tng100-1/output"
fields = ['SubhaloMass','SubhaloVelDisp']
subhalos = il.groupcat.loadSubhalos(basePath,99,fields=fields)

mass_msun = subhalos["SubhaloMass"]*1e10/0.704
plt.plot(mass_msun,subhalos['SubhaloVelDisp'],'.')

x = np.linspace(10**9,10**15,100)
y = x**0.25*0.1
plt.plot(x,y)

plt.xscale("log")
plt.yscale("log")
plt.xlabel("Total Mass [$M_\odot$]")
plt.ylabel("Velocity dispersion [(km/s)]")
plt.show()
```

The orange line represents the Faber-Jackson relation with constant $\gamma = 4$.
