```python
import numpy as np
import matplotlib.pyplot as plt

def Schechter(norm, alpha, char_lum, lum):
    return (norm*(lum/char_lum)**alpha*np.exp(-lum/char_lum)/char_lum)
```

```python
masses = np.logspace(8, 12, 100)
phi = Schechter(10**(22.4), 0.19, 10**11.2, masses)
fig1, axs1 = plt.subplots(nrows = 1, ncols = 1, figsize = (8,8))
axs1.plot(np.log10(masses), np.log10(phi))
```
